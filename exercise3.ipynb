```python
#4.3

#a) 
#rozwiązanie nieuniwersalne 
def iter_even(n):
    for i in range(0,n,2):
        yield i
#uniwersalne rozwiązanie:        
def iter_even_inf():
    i = 0
    while True:
        yield i
        i += 2
        
# drugi sposób dla znanego n: iter_even_2 = (i for i in range(0,n,2))

#b)
#w określonym zakresie:
def iter_odd(m):
    for i in range(1,m,2):
        yield i
        
#nieskończony:
def iter_odd_inf():
    i = 1
    while True:
        yield i
        i += 2
#c)       
def iter_power(k):
    i = 0
    while True:
        yield k**i
        i += 1
```
