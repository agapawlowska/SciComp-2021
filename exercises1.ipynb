```python
#zadanie 7
def roman2int(seq):
    k = 0       
    #tworzę słownik, uwzględniam wszystkie możliwe przypadki
    d1={"I":1, "IV":4, "V":5, "IX":9, "X":10, "XL":40, "L":50, "XC":90, "C":100, "CD":400,  "D":500, "CM":900, "M":1000}
   
    for i in range(len(seq)): #dla i w zakresie 0 do długosci liczby rzymskiej 
        if i > 0 and d1[seq[i]] > d1[seq[i-1]]: #dla i większego od 0 i liczby arabskiej odpowiadającej liczbie rzyskiej w seq sprawdzam warunek czy jest on większy od poprzedzającego wyrazu
            k += d1[seq[i]]         
        else:
            k += d1[seq[i]] # przeciwnym wypadku k zamieniam na liczbę przypadajacą i w słowniku
    return k
```

```python
roman2int('MCCIII')
```

```python
#zadanie 7, 2. metoda
def roman2int2(seq):
    d2={"I":1, "V":5, "X":10,  "L":50,  "C":100, "D":500, "M":1000}    
    seq = seq.replace("IV","IIII").replace('IX','I'*9).replace('XL','XXXX').replace('XC','X'*9).replace('CD','X'*4).replace('CM','C'*9)
    new_one = 0
    for i in seq:
        new_one += d2[i]
    return new_one
    
```
