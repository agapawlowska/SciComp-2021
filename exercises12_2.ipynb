```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
for node in (1, 2, 3, 4,5,6):
    G.add_node(node)
for (a, b) in [(1, 6), (1, 3), (2, 3), (3, 4), (4,5),(5,2),(6,3)]:
    G.add_edge(a, b)
    
nx.draw(G)
#nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
