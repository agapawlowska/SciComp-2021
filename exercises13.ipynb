```python
with Image.open("image.jpg") as img2:   
    print('mode: ', img2.mode)
    print('size: ', img2.size)
    print('width: ',img2.width)
    print('height: ',img2.height)
    print('format: ', img2.format)
    img2.thumbnail((128, 128))
    img2.save('makak_thumbnail.jpg', "JPEG")
```
