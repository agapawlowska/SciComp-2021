```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(1000000)
n = [10,100,1000,10000,100000,1000000]
in_circle = 0
outside_circle = 0

for j in n:
    for i in range(j):
        x = np.random.uniform(0,1)
        y = np.random.uniform(0,1)
        r = x**2 + y**2
        if r <= 1:
            in_circle += 1  
        outside_circle+= 1
        pi = 4*in_circle/outside_circle
    print(pi)
```
