```python
#zadanie 3.1
x = input("Enter a word: ")
a = ""
b = "+---"*len(x)+"+\n"

for i in x:
    a += f'|{i:^3}'

print(b + a+'|\n' + b)
```

```python
# zadanie 3.2 - for loop - wyprintowuję numer w () po to żeby widzieć czy pętla działa poprawnie

numbers = range(1,41)
for x in numbers:
    if x %5 ==0:
        print('10 is divided by 5')
    elif x%7 == 0:
        print('{} is divided by 7'.format(x))
    elif x == 13:
        pass
    else:
        print('({}) 3 is not important'.format(x))
```

```python
#zadanie 3.2 - while loop - korzystam również z for po to, aby iterować 

numbers = range(1,41)
for x in numbers:
    while x != 13:
        if x %5 ==0:
            print('10 is divided by 5')
            break
        elif x%7 == 0:
            print('{} is divided by 7'.format(x))
            break
        else:
            print('({}) 3 is not important'.format(x))
            break
        
    else:
        pass
```
