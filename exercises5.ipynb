```python
# EXERCISE 5.1

import os

def find_pdf_size(top):
    nr_of_bytes = 0
    for root, dirs, files in os.walk(top):   
        for name in files:
            if name.endswith(".pdf"):
                nr_of_bytes += os.path.getsize(os.path.join(root, name))
    return 'Sum of bytes taken by pdf files: {}'.format(nr_of_bytes)
```

```python
#EXERCISE 5.2

import datetime

def print_working_days(date1, date2):
    war1 = datetime.date.fromisoformat(date1)
    war2 = datetime.date.fromisoformat(date2)

    i = 0
    timedelta = war2 - war1
    for day in range(0,timedelta.days):
        if (war1 + datetime.timedelta(days=day)).isoweekday() <6:
            i += 1
            print(war1+ datetime.timedelta(days=day))
    print('{} working days in total'.format(i))
```

```python
print_working_days('2021-04-05', '2021-04-21')
```

```python
#EXERCISE 5.3
# nie mam pewności czy pozycja początkowa ma być ustalona jako 0 stąd też jako
# start przyjmuję jakikolwiek int, który jest moim punktem startowym, 
# następnie pozycja (x) zmienia się randomowo o +/-1 100-krotnie 
import random
random.seed()

def random_walk(start):
    n = 100
    x = start
    for i in range(0,n+1):
        x = random.choice([x-1, x+1])
    return x
```

```python
random_walk(0)
```

```python
random_walk(1)
```

```python
random_walk(-1)
```
