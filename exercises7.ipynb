```python
#EXERCISE 7.1
# Pierwsze podejście - bez wywoływania metody cross

class Vector:

    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        
    def __repr__(self):         
        return f"Vector({self.x},{self.y},{self.z})"
    
    def cross(self, other):  
        return Vector(self.y*other.z-self.z*other.y,self.z*other.x-self.x*other.z,self.x*other.y-self.y*other.x)
    
    def find_axis(self, other):
        if (self.y*other.z-self.z*other.y,self.z*other.x-self.x*other.z,self.x*other.y-self.y*other.x) == (0,0,0):
            raise ValueError("vectors are parallel")
        else:
            return self.cross(other)
```

```python
Vector.find_axis(Vector(1,2,3),Vector(2,4,6))
```

```python
Vector.find_axis(Vector(1,3,7),Vector(2,4,6))
```

```python
# Drugie podejście - z wykorzystaniem  metody cross (poprzez użycie self.metoda)
class Vector2:

    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        
    def __repr__(self):         
        return f"Vector({self.x},{self.y},{self.z})"
    
    def cross(self, other):  
        return Vector(self.y*other.z-self.z*other.y,self.z*other.x-self.x*other.z,self.x*other.y-self.y*other.x)
    
    def find_axis(self, other):
        if str(self.cross(other)) == str(Vector(0,0,0)):
            raise ValueError("vectors are parallel")
        else:
            return self.cross(other)
```

```python
Vector2.find_axis(Vector(1,9,7),Vector(2,4,6))
```

```python
Vector2.find_axis(Vector(1,2,3),Vector(2,4,6))
```

```python
#EXERCISE 7.2
import itertools
# (a) returning 0, 1, 0, 1, 0, 1, ...,
# pierwszy sposób
def iter1():
    d = True
    while d == True:
        n = 0
        yield n
        n = 1
        yield n
#drugi sposób
iter_zero_one = itertools.cycle([0,1])

# zwracanie sekwencji:
a = iter1()
seq1 = ''
for i in range(10):
    seq1 += str(next(a)) + ', '
print(seq1[:-2])        
```

```python
# (b) returning random sequence with 0 and 1
import random
random.seed()
# zwraca losowo 0 i 1 przy wywołaniu next(__)
def iter2():
    e = True
    while e == True:
        n = random.choice([0,1])
        yield n
```

```python
b = iter2()
next(b)
```

```python
seq2 = ''
for i in range(10):
    seq2 += str(next(b)) + ', '
print(seq2[:-2])
```
