```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# EXERCISE 9.1 (PLOT)
# Plot functions sin(x), cos(x), and exp(-x) in a range [0,10]. Colors are red, green, and blue, respectively. 
# Lines are solid, dashed, and dotted, respectively. Add a legend.

x = np.arange(0,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)
plt.plot(x,y1,color ='r')
plt.plot(x,y2,linestyle = 'dashed', color ='g')
plt.scatter(x,y3, color = 'b', s = 10)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['sin(x)', 'cos(x)','exp(-x)'])
plt.show()
```

```python
# EXERCISE 9.2 (SCATTER)
# Generate n=100 random points in a unit square [0,1]x[0,1]. Points are green if the distance from (0,0) is less then 1; 
# they are red otherwise. The area of a point (x,y) should be proportional to x+y.
np.random.seed(1000000)
n = 100

x = np.random.rand(n)
y = np.random.rand(n)
#area przemnaam przez 25 eby lepiej uwidoczni zaleno rozmiaru od x+y
area = 25*(x+y)
colors = ['g' if cond else 'r' for cond in (x**2+y**2<1)]
plt.scatter(x,y,area,colors)
plt.show()
```
